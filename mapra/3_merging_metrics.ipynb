```python
from mapra import prep
from pathlib import Path
import plotly.express as px
data = prep.dataset(Path('.').resolve().parent)
df = data.dataframe_remerged()
fig = px.scatter_3d(df, x='dtemp', y='ddg', z='h2o', color='desc', width=800, height=800, template='simple_white')
fig.update_traces(mode='markers', marker_line_width=0, marker_size=4, opacity=.5)

```

```python
df = data.dataframe_remerged(df=data.dataframe_repeats_avg())
fig = px.scatter_3d(df, x='dtemp', y='ddg', z='h2o', color='desc', width=800, height=800, title='simple averaging', template='simple_white')
fig.update_traces(mode='markers', marker_line_width=0, marker_size=4, opacity=.5)
```

```python
df = data.dataframe_remerged(df=data.dataframe_gaussian_avg())
fig = px.scatter_3d(df, x='dtemp', y='ddg', z='h2o', color='desc', width=800, height=800, title='weighted (Gaussian) averaging', template='simple_white')
fig.update_traces(mode='markers', marker_line_width=0, marker_size=4, opacity=.5)
```
